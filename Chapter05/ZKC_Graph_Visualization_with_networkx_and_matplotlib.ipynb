# Visualizing a graph without node's known positions

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Karate club graph

```python
G = nx.karate_club_graph()
```

```python
type(G), G.number_of_nodes(), G.number_of_edges()
```

```python
list(G.nodes(data=True))[:2]
```

```python
list(G.edges(data=True))[:2]
```

```python
G.nodes.data("club")
```

```python
pos = nx.random_layout(G, seed=42)
pos
```

```python
nx.draw(G, pos=pos)
```

```python
pos = nx.circular_layout(G)
nx.draw(G, pos=pos)
```

```python
pos = nx.kamada_kawai_layout(G)
nx.draw(G, pos=pos)
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos=pos)
```

## Configuring node and edge display

```python
nx.draw_spring(
    G, 
    node_size=150,
    node_color=["black" if node["club"] == "Mr. Hi" else "grey" for _, node in G.nodes(data=True)],
    width=[e["weight"]/2. for _, _, e in G.edges(data=True)]
)
```
